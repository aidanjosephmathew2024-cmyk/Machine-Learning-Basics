```python
df['social_media_hours'] = df['social_media_hours'].fillna(df['social_media_hours'].median())
```

```python
df['focus_score'] = df['focus_score'].fillna(df['focus_score'].median())
```

```python
df['sleep_hours'] = df['sleep_hours'].fillna(df['sleep_hours'].median())
```

```python
print(df.isnull().sum())
```

```python
numerical_col = df.select_dtypes(include=['int64', 'float64']).columns
numerical_col
```

```python
categorical_col = df.select_dtypes(include=['object']).columns
categorical_col
```

```python
for col in numerical_col:
  plt.figure(figsize=(10,6))
  sns.boxplot(df[col])
  plt.title(col)
  plt.show()
```

```python
for col in numerical_col:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3-Q1
  LL = Q1-1.5*IQR
  UL = Q3+1.5*IQR
  df[col] = np.where(df[col]>UL,UL,df[col])
  df[col] = np.where(df[col]<LL, LL, df[col])
```

```python
for col in numerical_col:
  plt.figure(figsize=(10,6))
  sns.boxplot(df[col])
  plt.title(col)
  plt.show()
```

```python
categorical_col = df.select_dtypes(include=['object']).columns
categorical_col
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in categorical_col:
  df[col] = le.fit_transform(df[col])
df.head()
```

```python
df.drop('student_id', axis=1, inplace=True)
```

```python
df.head()
```

```python
x = df.drop('performance_category', axis=1)
y = df['performance_category']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state = 42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.show()
```

```python
#To check if its overfitting or underfitting
train_accuracy = model.score(x_train, y_train)
test_accuracy = model.score(x_test, y_test)
print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
```
